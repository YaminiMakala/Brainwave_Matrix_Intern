```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('train.csv', parse_dates=['Order Date', 'Ship Date'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['Postal Code'])
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day
df['Order Weekday'] = df['Order Date'].dt.weekday
```

```python
print(df[['Order Date', 'Order Year', 'Order Month', 'Order Day', 'Order Weekday']].head())
```

Exploratory Data Analysis (EDA)
a. Sales Over Time

```python
# Plot sales over months
monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Order Month', y='Sales', hue='Order Year')
plt.title('Monthly Sales Over Years')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

b. Top Selling Products

```python
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar')
plt.title('Top 10 Selling Products')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

c. Sales by Category and Sub-Category

```python
category_sales = df.groupby('Category')['Sales'].sum()
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().nlargest(10)
```

```python
# Category Sales
plt.figure(figsize=(8, 4))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Sales by Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Sub-Category Sales
plt.figure(figsize=(12, 6))
subcategory_sales.plot(kind='bar', color='orange')
plt.title('Top 10 Sub-Category Sales')
plt.ylabel('Total Sales')
plt.show()
```

d. Profit vs. Sales

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Sales', y='Sales', hue='Category')  # Using 'Sales' on both axes for demonstration
plt.title('Sales Distribution by Category')
plt.xlabel('Sales')
plt.ylabel('Sales')
plt.show()
```

```python
from prophet import Prophet
```

```python
# Aggregate daily sales
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
# Rename columns for Prophet
daily_sales.rename(columns={'Order Date': 'ds', 'Sales': 'y'}, inplace=True)
```

```python
# Use data except the last 7 days for training
train_data = daily_sales[:-7]
test_data = daily_sales[-7:]
```

```python
model = Prophet()
model.fit(train_data)
```

```python
# Create a dataframe to hold predictions
future = model.make_future_dataframe(periods=7)
forecast = model.predict(future)
```

```python
model.plot(forecast)
plt.title('Sales Forecast for Next 7 Days')
plt.show()
```

```python
# Compare predicted vs actual sales
forecast_actual = forecast[['ds', 'yhat']].merge(test_data, on='ds')
print(forecast_actual[['ds', 'yhat', 'y']])
```
